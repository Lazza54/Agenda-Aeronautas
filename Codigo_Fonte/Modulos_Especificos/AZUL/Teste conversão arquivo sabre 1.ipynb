```python
# RE para este formato de data 31 May 2025
import re
# Exemplo de regex para datas no formato 31 May 2025
p0 = re.compile(r'(\d{1,2} [A-Za-z]{3} \d{4})')  # Exemplo de regex para datas no formato 31 May 2025
texto_formatado1 = texto_completo  # O texto já está formatado pela função processar_texto
# Encontra todas as datas no texto formatado
datas_encontradas = p0.findall(texto_formatado1)  # Encontra todas as datas no texto formatado
print("Datas encontradas:")
for data in datas_encontradas:
    print(data)  # Exibe cada data encontrada
    
# Exemplo de regex para horas no formato HH:MM
p2 = re.compile(r'(\d{2}:\d{2})')  # Exemplo de regex para horas no formato HH:MM
horas = p2.findall(texto_formatado1)  # Encontra todas as horas no texto formatado
print("Horas encontradas:")
for hora in horas:
    print(hora)  # Exibe cada hora encontrada
    
# Exemplo de regex para encontrar Apresentação
p3 = re.compile(r'Apresentação\s+(\d{1,2}:\d{2})')  # Exemplo de regex para encontrar "Apresentação HH:MM"
apresentacoes = p3.findall(texto_formatado1)  # Encontra todas as apresentações no texto formatado
print("Apresentações encontradas:")
for apresentacao in apresentacoes:
    print(apresentacao)  # Exibe cada apresentação encontrada
       
# Exemplo de regex para encontrar string que inicie com 'AD'"ARQUIVOS COMUNS"
p4 = re.compile(r'\bAD\w*')  # Exemplo de regex para encontrar palavras que começam com 'AD'
ad_strings = p4.findall(texto_formatado1)  # Encontra todas as strings que começam com 'AD' no texto formatado
print("Strings que começam com 'AD':")
for ad_string in ad_strings:
    print(ad_string)  # Exibe cada string que começa com 'AD'
    
# substituir todos os espaços por virgulas
texto_formatado1 = texto_formatado1.replace('  ', ',')  # Substitui todos os espaços por vírgulas
print("Texto formatado com espaços substituídos por vírgulas:")
print(texto_formatado1)  # Exibe o texto formatado com espaços substituídos por vírgulas
# Exemplo de regex para encontrar palavras que começam com 'AD' e terminam com 'ARQUIVOS COMUNS'

# extrair do texto todas as palavras que iniciam em maiuscula e tenham uma virgula no final
import re
p5 = re.compile(r'\b[A-Z][a-zA-Z]*,')  # Exemplo de regex para encontrar palavras que começam com maiúscula e terminam com vírgula
palavras_maiusculas = p5.findall(texto_formatado1)  # Encontra todas as palavras que começam com maiúscula e terminam com vírgula no texto formatado
print("Palavras que começam com maiúscula e terminam com vírgula:")
for palavra in palavras_maiusculas:
    print(palavra)  # Exibe cada palavra que começa com maiúscula e termina com vírgula
    
# Exemplo de regex para horas no formato HH:MM da primeira linha do texto formatado
import re
p15 = re.compile(r'(\d{2}:\d{2})')  # Exemplo de regex para horas no formato HH:MM
texto_formatado1 = texto_completo  # O texto já está formatado pela função processar_texto
horas_encontradas = p15.findall(texto_formatado1)  # Encontra todas as horas no texto formatado
print("Horas encontradas na primeira linha:")
# buscar na primeira linha do texto formatado a data e hora
if horas_encontradas:
    primeira_linha = horas_encontradas[1]  # Pega a primeira hora encontrada
    print(primeira_linha)  # Exibe a primeira hora encontrada
        

        
```
