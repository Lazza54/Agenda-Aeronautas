ESTE CÓDIGO UTILIZARÁ O ARQUIVO PDF DA ESCALA MENSAL PUBLICADA NO SITE SABRE DA EMPRESA E CRIARÁ UM DATAFRAME PANDAS PARA REALIZAR AS CORREÇÕES NECESSÁRIAS, SALVARÁ EM UM ARQUIVO .CSV PARA USO POSTERIOR

```python
# PASSO 1: IMPORTAR O ARQUIVO .PDF RECEBIDO DO SABRE
# IMPORTAR AS BIBLIOTECAS NECESSÁRIAS PARA A IMPORTAÇÃO
import pandas as pd 
import PyPDF2

# IMPORTAR AS BIBLIOTECAS NECESSÁRIAS PARA A ESCOLHA DO ARQUIVOS filedialog
from tkinter import filedialog

# CONFIGURAR PARA QUE NÃO APARECAM AS LINHAS AZUIS ABAIXO DAS PALAVRAS NO EDITOR 
pd.options.mode.chained_assignment = None  # default='warn'

# IGNORAR AVISOS DO PANDAS
import warnings
warnings.filterwarnings("ignore", category=UserWarning, module='pandas')

# ABRIR O DIRETÓRIO E ESCOLHER O ARQUIVO .PDF janela para escolha deve sobrepor as demais janelas
# SOBREPOR AS DEMAIS JANELAS
from tkinter import Tk
root = Tk()
root.withdraw()
root.lift()
root.attributes('-topmost', True)

path = filedialog.askopenfilename(filetypes=[("PDF files", "*.pdf")])

print(path)

if path.lower().endswith('.pdf'):
	imported_data = open(path, 'rb')
	pdf_reader = PyPDF2.PdfReader(imported_data)
	num_paginas = len(pdf_reader.pages)
	print(f"Número de páginas: {num_paginas}")
else:
	print("Por favor, selecione um arquivo PDF válido.")

# PASSO 2: ANALISE PRÉVIA DO ARQUIVO
display(imported_data)
# EXIBIR O CONTEÚDO DO ARQUIVO PDF
for i in range(num_paginas):
    page = pdf_reader.pages[i]
    print(f"Conteúdo da página {i + 1}:")
    print(page.extract_text())
```

 *** AQUI VAMOS TRABALHAR AS FOLGAS

```python
import re

# PASSO 3: AÇÕES A SEREM TOMADAS APÓS A ANALISE PRÉVIA
# AQUI VOCÊ PODE ADICIONAR CÓDIGOS PARA REALIZAR AÇÕES ESPECÍFICAS
# SALVAR A PRIMEIRA LINHA NA VARIÁVEL QUE SERÁ O NOME DO ARQUIVO .CSV AO FINAL DO CÓDIGO
# IR PARA O INÍCIO DO ARQUIVO E EXTRAIR A PRIMEIRA LINHA
page = pdf_reader.pages[0]
nome_do_arquivo = page.extract_text().split('\n')[0]
print(nome_do_arquivo)

# FILTRAR O ARQUIVO ONDE AS LINHAS SE INICIEM COM 'Off'"ARQUIVOS COMUNS"
linhas_filtradas_Off = []
for i in range(num_paginas):
    page = pdf_reader.pages[i]
    texto = page.extract_text()
    if texto:
        linhas = texto.split('\n')
        for linha in linhas:
            if linha.startswith('Off'):
                linhas_filtradas_Off.append(linha)
display(linhas_filtradas_Off)                

# INCLUIR NAS LINHAS DE OS VALORES DAS LINHAS IMEDIATAMENTE ANTERIORES A ELAS
# Para garantir que a inclusão das linhas anteriores ocorra desde o início do arquivo,
# a linha a ser incluída deve ser a que começar com o nome do dia da semana (3 letras em inglês), seguido de uma vírgula.

linhas_filtradas_Off_com_valores = []
linha_dia_semana = None
padrao_dia_semana = re.compile(r'^(Mon|Tue|Wed|Thu|Fri|Sat|Sun),')

for i in range(num_paginas):
  page = pdf_reader.pages[i]
  texto = page.extract_text()
  if texto:
    linhas_pagina = texto.split('\n')
    for linha in linhas_pagina:
      if padrao_dia_semana.match(linha):
        linha_dia_semana = linha
      if linha.startswith('Off'):
        if linha_dia_semana:
          linhas_filtradas_Off_com_valores.append(f"{linha_dia_semana}, {linha}")
        else:
          linhas_filtradas_Off_com_valores.append(linha)
display(linhas_filtradas_Off_com_valores)


##############################################


# FILTRAR O ARQUIVO ONDE AS LINHAS SE INICIEM COM 'Reserva'"ARQUIVOS COMUNS"
linhas_filtradas_Reserva = []
for i in range(num_paginas):
    page = pdf_reader.pages[i]
    texto = page.extract_text()
    if texto:
        linhas = texto.split('\n')
        for linha in linhas:
            if linha.startswith('Reserva'):
                linhas_filtradas_Reserva.append(linha)
display(linhas_filtradas_Reserva)                

# INCLUIR NAS LINHAS DE OS VALORES DAS LINHAS IMEDIATAMENTE ANTERIORES A ELAS
# Para garantir que a inclusão das linhas anteriores ocorra desde o início do arquivo,
# a linha a ser incluída deve ser a que começar com o nome do dia da semana (3 letras em inglês), seguido de uma vírgula.

linhas_filtradas_Reserva_com_valores = []
linha_dia_semana = None
padrao_dia_semana = re.compile(r'^(Mon|Tue|Wed|Thu|Fri|Sat|Sun),')

for i in range(num_paginas):
  page = pdf_reader.pages[i]
  texto = page.extract_text()
  if texto:
    linhas_pagina = texto.split('\n')
    for linha in linhas_pagina:
      if padrao_dia_semana.match(linha):
        linha_dia_semana = linha
      if linha.startswith('Reserva'):
        if linha_dia_semana:
          linhas_filtradas_Reserva_com_valores.append(f"{linha_dia_semana}, {linha}")
        else:
          linhas_filtradas_Reserva_com_valores.append(linha)
display(linhas_filtradas_Reserva_com_valores)

####################################### 


# FILTRAR O ARQUIVO ONDE AS LINHAS SE INICIEM COM 'AD' VOOS REALIZADOS
linhas_filtradas_Voos = []
for i in range(num_paginas):
    page = pdf_reader.pages[i]
    texto = page.extract_text()
    if texto:
        linhas = texto.split('\n')
        for linha in linhas:
            if linha.startswith('AD'):
                linhas_filtradas_Voos.append(linha)
display(linhas_filtradas_Voos)                

# INCLUIR NAS LINHAS DE OS VALORES DAS LINHAS IMEDIATAMENTE ANTERIORES A ELAS
# Para garantir que a inclusão das linhas anteriores ocorra desde o início do arquivo,
# a linha a ser incluída deve ser a que começar com o nome do dia da semana (3 letras em inglês), seguido de uma vírgula.

linhas_filtradas_Voos_com_valores = []
linha_dia_semana = None
padrao_dia_semana = re.compile(r'^(Mon|Tue|Wed|Thu|Fri|Sat|Sun),')

for i in range(num_paginas):
  page = pdf_reader.pages[i]
  texto = page.extract_text()
  if texto:
    linhas_pagina = texto.split('\n')
    for linha in linhas_pagina:
      if padrao_dia_semana.match(linha):
        linha_dia_semana = linha
      if linha.startswith('AD'):
        if linha_dia_semana:
          linhas_filtradas_Voos_com_valores.append(f"{linha_dia_semana}, {linha}")
        else:
          linhas_filtradas_Voos_com_valores.append(linha)
display(linhas_filtradas_Voos_com_valores)
```

AQUI VAMOS TRABALHAR AS FOLGAS

```python
# STEP 4:
# CREATE A DATAFRAME WITH THE FOLLOWING COLUMNS:
# 'Activity' 'Checkin' 'Start' 'Dep' 'Arr' 'End' 'Checkout' 'Daytime' 'NightTime' 'TotalTime' 'AcVer' 'Prefix' 'DD' 'CAT' 'Crew'

df_primary = pd.DataFrame(columns=['Activity', 'Checkin', 'Start', 'Dep', 'Arr', 'End', 'Checkout', 'Daytime', 'NightTime', 'TotalTime', 'AcVer', 'Prefix', 'DD', 'CAT', 'Crew'])

# ADD THE FILTERED 'Off' LINES TO THE DATAFRAME
for line in linhas_filtradas_Off_com_valores:
  parts = line.split(',')
  if len(parts) >= 2:
    # Obter os caracteres 5 e 6 do texto

    # Extrai o conteúdo do segundo grupo entre parênteses, se existir
    conteudos = []
    matches = re.findall(r'\((.*?)\)', line)
    if len(matches) >= 2:
      conteudos = matches[1]
    elif matches:
      conteudos = matches[0]
    
    # Pega o texto após a primeira vírgula até o primeiro parêntese
    after_first_comma = line.split(',', 1)[1] if ',' in line else ''
    before_first_parenthesis = after_first_comma.split('(', 1)[0].strip()
    #print(f"Conteúdo após a primeira vírgula e antes do primeiro parêntese: {before_first_parenthesis}")

    # CONVERTER O VALOR QUE ESTÁ EM before_first_parenthesis SENDO dd mm yyyy para o formato dd/mm/yyyy 
    match_data = re.search(r'(\d{1,2})\s+(\w+)\s+(\d{4})', before_first_parenthesis)
    if match_data:
        dia, mes, ano = match_data.groups()
        meses = {
            'Jan': '01', 'January': '01', 'Feb': '02', 'February': '02', 'Mar': '03', 'March': '03', 'Apr': '04', 'April': '04', 'Apr': '04',
            'May': '05', 'Jun': '06', 'June': '06', 'Jul': '07', 'July': '07', 'Aug': '08', 'August': '08',
            'Sep': '09', 'September': '09', 'Oct': '10',  'October': '10', 'Nov': '11',  'november': '11', 'Dec': '12', 'December': '12'
        }
        mes_num = meses.get(mes, mes)
        data_formatada = f"{dia.zfill(2)}/{mes_num}/{ano}"
        #print(f"Data formatada: {data_formatada}")
    else:
        data_formatada = ''
        print("Data não encontrada ou no formato inesperado.")
        
  ##########################  
    
    # Busca o primeiro horário no formato ??h?? ou ??:??
    match_hora = re.search(r'\b\d{2}:\d{2}\b', line)
    hora = match_hora.group(0) if match_hora else ''
    #print(f"Horário encontrado: {hora}")  
    
    # JUNTAR O CONTEÚDO DO match_hora COM A DATA FORMATADA
    if data_formatada and hora:
        match_hora = f"{data_formatada} {hora}"
    else:
        match_hora = hora or data_formatada
    #print(f"Horário combinado: {match_hora}")

    # COLOCAR NA VARIÁVEL var_localidade os últimos 3 caracteres do texto
    var_localidade = line[-3:] if len(line) >= 3 else ''
    #print(f"Localidade: {var_localidade}")
    
    # CRIAR A VARIÁVEL PARA O End e Checkout
    # Se a linha contiver 'Off' ou 'Reserva', o End e Checkout serão iguais ao Checkin + 1 dia
    if 'Off' in line or 'Reserva' in line:
        # Incrementar a data em 1 dia
        match_hora_parts = match_hora.split()
        if len(match_hora_parts) == 2:
            data_part, hora_part = match_hora_parts
            dia, mes, ano = map(int, data_part.split('/'))
            from datetime import datetime, timedelta
            data_obj = datetime.strptime(f"{ano}-{mes}-{dia}", "%Y-%m-%d") + timedelta(days=1)
            end_checkout = data_obj.strftime("%d/%m/%Y") + ' ' + hora_part
        else:
            end_checkout = match_hora  # Se não houver data, manter o checkin
    else:
        end_checkout = ''  # Fill as needed
    

    new_row = {
  
      'Activity': conteudos,
      'Checkin': match_hora,
      'Start': match_hora,
      'Dep': var_localidade,
      'Arr': var_localidade,
      'End': end_checkout,
      'Checkout': end_checkout,
      'Daytime': '',  # Fill as needed
      'NightTime': '',  # Fill as needed
      'TotalTime': '',  # Fill as needed
      'AcVer': '',  # Fill as needed
      'Prefix': '',  # Fill as needed
      'DD': '',  # Fill as needed
      'CAT': '',  # Fill as needed
      'Crew': ''   # Fill as needed
    }
    df_primary = pd.concat([df_primary, pd.DataFrame([new_row])], ignore_index=True)

display(df_primary)
```

```python
# ORGANIZAR O DATAFRAME POR Start
# 🔧 CORREÇÃO: Converter Start para datetime apenas temporariamente para ordenação
print("🚀 INICIANDO ORGANIZAÇÃO E CORREÇÃO DO DATAFRAME...")

# 🔍 DIAGNÓSTICO INICIAL: Verificar estado atual das colunas
print("\n🔍 DIAGNÓSTICO INICIAL:")
for col in ['Checkin', 'Start', 'End', 'Checkout']:
    if col in df_primary.columns:
        tipo_atual = df_primary[col].dtype
        valores_vazios = df_primary[col].astype(str).str.strip().eq('').sum()
        valores_na = df_primary[col].isna().sum()
        print(f"   • {col}: Tipo={tipo_atual}, Vazios={valores_vazios}, NaN={valores_na}")

# 🔧 CORREÇÃO: Se Start estiver vazia, copiar de Checkin
if df_primary['Start'].isna().all() or df_primary['Start'].astype(str).str.strip().eq('').all():
    print("\n⚠️ PROBLEMA DETECTADO: Coluna Start está vazia - copiando de Checkin...")
    df_primary['Start'] = df_primary['Checkin'].copy()
    print("✅ Coluna Start reconstruída a partir de Checkin")

# 🔧 ORDENAÇÃO: Criar coluna temporária para ordenar
print("\n📊 ORDENANDO DATAFRAME POR DATA/HORA...")
df_primary_temp = df_primary.copy()

# Converter Start para datetime temporariamente para ordenação
df_primary_temp['Start_temp'] = pd.to_datetime(df_primary_temp['Start'], format='%d/%m/%Y %H:%M', errors='coerce')

# Ordenar pelo campo temporário
df_primary_temp.sort_values(by='Start_temp', inplace=True)

# Remover a coluna temporária e manter a coluna Start original
df_primary = df_primary_temp.drop('Start_temp', axis=1)

# REINDEXAR O DATAFRAME
df_primary.reset_index(drop=True, inplace=True)
print("✅ DataFrame ordenado e reindexado")

# 🔧 VERIFICAÇÃO E CORREÇÃO DE FORMATO: Garantir que todas as colunas de data estão em formato string
print("\n🔧 VERIFICAÇÃO E CORREÇÃO DE FORMATO DAS COLUNAS:")
for col in ['Checkin', 'Start', 'End', 'Checkout']:
    if col in df_primary.columns:
        print(f"\n   🔍 Processando coluna: {col}")
        
        # Verificar o tipo atual
        tipo_atual = df_primary[col].dtype
        print(f"      • Tipo atual: {tipo_atual}")
        
        # Se for datetime, converter para string
        if 'datetime' in str(tipo_atual):
            print(f"      ⚠️ Convertendo {col} de datetime para string...")
            df_primary[col] = df_primary[col].dt.strftime('%d/%m/%Y %H:%M')
            print(f"      ✅ {col} convertido para string")
        
        # Se for object, verificar e limpar valores problemáticos
        elif tipo_atual == 'object':
            # Verificar se há valores NaN ou vazios
            valores_na = df_primary[col].isna().sum()
            if valores_na > 0:
                print(f"      ⚠️ {col} tem {valores_na} valores NaN - substituindo por string vazia...")
                df_primary[col] = df_primary[col].fillna('')
                print(f"      ✅ {col} - valores NaN corrigidos")
            
            # Verificar se há valores com padding desnecessário
            if df_primary[col].astype(str).str.contains(r'\s+$').any():
                print(f"      ⚠️ {col} tem espaços em branco - removendo padding...")
                df_primary[col] = df_primary[col].astype(str).str.strip()
                print(f"      ✅ {col} - padding removido")
        
        # Mostrar exemplos após correção
        exemplos = df_primary[col].head(3).tolist()
        print(f"      📋 Exemplos após correção: {exemplos}")

print("\n✅ Verificação e correção de formato concluída!")

# 🔧 CORREÇÃO: Extrair o diretório do caminho do arquivo PDF
import os

# Extrair apenas o diretório do path (remover o nome do arquivo)
diretorio_original = os.path.dirname(path)
print(f"\n📁 Diretório original: {diretorio_original}")

# Mudar para o diretório onde está o arquivo PDF
os.chdir(diretorio_original)
print(f"📁 Diretório atual: {os.getcwd()}")

# Criar o nome do arquivo CSV
nome_do_arquivo_csv = f"{nome_do_arquivo}.csv"

# 🔍 VERIFICAÇÃO FINAL: Mostrar estatísticas antes de salvar
print(f"\n📋 VERIFICAÇÃO FINAL ANTES DE SALVAR:")
print(f"   • Total de linhas: {len(df_primary)}")
print(f"   • Total de colunas: {len(df_primary.columns)}")
print(f"   • Colunas: {list(df_primary.columns)}")

# Verificar especificamente as colunas críticas
for col in ['Activity', 'Checkin', 'Start', 'End', 'Checkout']:
    if col in df_primary.columns:
        valores_vazios = df_primary[col].astype(str).str.strip().eq('').sum()
        valores_na = df_primary[col].isna().sum()
        exemplos = df_primary[col].head(3).tolist()
        print(f"\n   📊 Estatísticas da coluna '{col}':")
        print(f"      • Exemplos: {exemplos}")
        print(f"      • Valores vazios: {valores_vazios}")
        print(f"      • Valores NaN: {valores_na}")
        print(f"      • Tipo de dados: {df_primary[col].dtype}")

# 💾 SALVAR O DATAFRAME COMO UM ARQUIVO .CSV
print(f"\n💾 SALVANDO ARQUIVO CSV...")
try:
    df_primary.to_csv(nome_do_arquivo_csv, index=False)
    print(f"✅ Arquivo CSV salvo com sucesso: {nome_do_arquivo_csv}")
    
    # Mostrar o caminho completo do arquivo salvo
    caminho_completo_csv = os.path.join(diretorio_original, nome_do_arquivo_csv)
    print(f"📁 Caminho completo: {caminho_completo_csv}")
    
    # Verificar se o arquivo foi realmente criado
    if os.path.exists(caminho_completo_csv):
        tamanho_arquivo = os.path.getsize(caminho_completo_csv)
        print(f"📏 Tamanho do arquivo: {tamanho_arquivo} bytes")
    else:
        print("⚠️ AVISO: Arquivo não foi encontrado após salvamento")
        
except Exception as e:
    print(f"❌ ERRO ao salvar arquivo: {e}")

# 📊 EXIBIR O DATAFRAME FINAL
print(f"\n📊 DATAFRAME FINAL:")
print(f"   Primeiras 5 linhas:")
display(df_primary.head(50))

print(f"\n🎉 PROCESSO CONCLUÍDO COM SUCESSO!")
print(f"   ✅ DataFrame organizado por data")
print(f"   ✅ Colunas de data em formato string correto")
print(f"   ✅ Arquivo CSV salvo")
print(f"   ✅ Todas as verificações passaram")
```
