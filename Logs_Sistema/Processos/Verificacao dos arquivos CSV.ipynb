```python
# IMPORTAR BIBLIOTECAS necess√°rias
from datetime import datetime, time

import pandas as pd
import numpy as np
```

```python
def escolher_diretorio():
    from tkinter import Tk
    from tkinter.filedialog import askdirectory

    # criar uma janela oculta
    Tk().withdraw()
    # abrir o dialogo para escolher o diretorio
    diretorio = askdirectory(title="Escolha o diret√≥rio")
    return diretorio

# criar uma fun√ß√£o para escolher os arquivos csv
def escolher_arquivos():
    from tkinter import Tk
    from tkinter.filedialog import askopenfilenames

    # criar uma janela oculta
    Tk().withdraw()
    # abrir o dialogo para escolher os arquivos
    arquivos = askopenfilenames(title="Escolha os arquivos CSV", filetypes=[("CSV files", "*.csv")])
    return arquivos if arquivos else None
```

```python
# chamar a fun√ß√£o para escolher o diretorio
diretorio = escolher_diretorio()
if diretorio:
    print(f"‚úÖ Diret√≥rio selecionado: {diretorio}")
else:
    print("‚ö†Ô∏è Nenhum diret√≥rio foi selecionado")

# chamar a fun√ß√£o para escolher os arquivos csv
arquivos = escolher_arquivos()

# verificar se os arquivos foram escolhidos
if arquivos is None or len(arquivos) == 0:
    print("‚ùå Nenhum arquivo foi selecionado.")
    print("üí° Dica: Execute a c√©lula novamente e selecione exatamente 2 arquivos CSV")
elif len(arquivos) != 2:
    print(f"‚ö†Ô∏è Foram selecionados {len(arquivos)} arquivos. Por favor, selecione exatamente 2 arquivos CSV.")
    print("üí° Execute a c√©lula novamente e selecione apenas 2 arquivos")
else:
    print(f"‚úÖ Exatos 2 arquivos selecionados!")
    
    # mostrar os arquivos selecionados
    arquivo1 = arquivos[0]
    arquivo2 = arquivos[1]
    print(f"1. {arquivo1.split('\\')[-1]}")
    print(f"2. {arquivo2.split('\\')[-1]}")
    
    print("\nüìÇ Carregando os arquivos...")
    
    try:
        # carregar o primeiro arquivo
        df1 = pd.read_csv(arquivo1, sep=',', encoding='utf-8')
        nome_arquivo1 = arquivo1.split('\\')[-1]
        print(f"‚úÖ Arquivo 1 carregado: {nome_arquivo1} - Shape: {df1.shape}")
        
        # carregar o segundo arquivo
        df2 = pd.read_csv(arquivo2, sep=',', encoding='utf-8')
        nome_arquivo2 = arquivo2.split('\\')[-1]
        print(f"‚úÖ Arquivo 2 carregado: {nome_arquivo2} - Shape: {df2.shape}")
        
        print("\n" + "="*50)
        print(f"üìä DISPLAY DO PRIMEIRO ARQUIVO: {nome_arquivo1}")
        print("="*50)
        print(f"üîπ Colunas: {list(df1.columns)}")
        print(f"üîπ Primeiras 5 linhas:")
        print(df1.head())
        
        print("\n" + "="*50)
        print(f"üìä DISPLAY DO SEGUNDO ARQUIVO: {nome_arquivo2}")
        print("="*50)
        print(f"üîπ Colunas: {list(df2.columns)}")
        print(f"üîπ Primeiras 5 linhas:")
        print(df2.head())
        
        # informa√ß√µes detalhadas usando info()
        print("\nüìä INFORMA√á√ïES DETALHADAS DO ARQUIVO 1:")
        print(f"üîπ {nome_arquivo1}")
        df1.info()
        
        print("\nüìä INFORMA√á√ïES DETALHADAS DO ARQUIVO 2:")
        print(f"üîπ {nome_arquivo2}")
        df2.info()
        
        # resumo r√°pido da compara√ß√£o
        print(f"\nüîÑ RESUMO COMPARATIVO:")
        print(f"üìã Arquivo 1: {df1.shape[0]:,} linhas x {df1.shape[1]} colunas")
        print(f"üìã Arquivo 2: {df2.shape[0]:,} linhas x {df2.shape[1]} colunas")
        print(f"üìä Diferen√ßa: {abs(df1.shape[0] - df2.shape[0]):,} linhas, {abs(df1.shape[1] - df2.shape[1])} colunas")
        
        print("\n‚úÖ PROCESSAMENTO CONCLU√çDO!")
        
    except Exception as e:
        print(f"‚ùå Erro ao carregar os arquivos: {e}")
```

INICIAR O LOOP PARA COMPARA√á√ÉO DAS COLUNAS EM AMBOS ARQUIVOS

```python
#df1.info()
#df2.info()

print("üîÑ COMPARA√á√ÉO BASEADA NO ARQUIVO BASE (df1)")
print("="*60)

# Obter o n√∫mero de colunas do df1 (arquivo base)
num_colunas_base = len(df1.columns)
print(f"üìä N√∫mero de colunas do arquivo base (df1): {num_colunas_base}")
print(f"üìä N√∫mero de colunas do arquivo calculado (df2): {len(df2.columns)}")

# Extrair as primeiras N colunas do df2 correspondentes ao df1
colunas_df1 = list(df1.columns)
colunas_df2_base = list(df2.columns[:num_colunas_base])

print(f"\nüîç COMPARA√á√ÉO DAS COLUNAS BASE:")
print(f"üîπ Colunas do arquivo 1: {colunas_df1}")
print(f"üîπ Primeiras {num_colunas_base} colunas do arquivo 2: {colunas_df2_base}")

# Verificar se as colunas base s√£o iguais
if colunas_df1 == colunas_df2_base:
    print("‚úÖ As colunas base dos dois arquivos s√£o iguais.")
    colunas_base_iguais = True
else:
    print("‚ö†Ô∏è As colunas base dos dois arquivos s√£o diferentes.")
    colunas_base_iguais = False
    
    # Mostrar diferen√ßas espec√≠ficas
    for i, (col1, col2) in enumerate(zip(colunas_df1, colunas_df2_base)):
        if col1 != col2:
            print(f"   üî∏ Posi√ß√£o {i}: '{col1}' vs '{col2}'")

# Extrair apenas as colunas base do df2 para compara√ß√£o
df2_base = df2.iloc[:, :num_colunas_base].copy()
df2_base.columns = colunas_df1  # Garantir que os nomes das colunas sejam iguais

print(f"\nüîç COMPARA√á√ÉO DOS TIPOS DE DADOS BASE:")
# Verificar se os tipos de dados das colunas base s√£o iguais
tipos_df1 = df1.dtypes
tipos_df2_base = df2_base.dtypes

if tipos_df1.equals(tipos_df2_base):
    print("‚úÖ Os tipos de dados das colunas base s√£o iguais.")
    tipos_iguais = True
else:
    print("‚ö†Ô∏è Os tipos de dados das colunas base s√£o diferentes.")
    tipos_iguais = False
    
    # Mostrar diferen√ßas nos tipos
    for coluna in colunas_df1:
        if tipos_df1[coluna] != tipos_df2_base[coluna]:
            print(f"   üî∏ Coluna '{coluna}': {tipos_df1[coluna]} vs {tipos_df2_base[coluna]}")

# Adicionar esta se√ß√£o ap√≥s a compara√ß√£o dos tipos de dados e antes da an√°lise de valores nulos

print(f"\nüîç VERIFICA√á√ÉO DETALHADA DO CONTE√öDO DAS COLUNAS BASE:")
print("="*60)

# Verificar se o conte√∫do das primeiras 12 colunas √© id√™ntico
conteudo_identico = True
diferencas_por_coluna = {}

for coluna in colunas_df1:
    print(f"\nüìã Analisando coluna: '{coluna}'")
    
    # Comparar valores da coluna
    coluna_df1 = df1[coluna]
    coluna_df2_base = df2_base[coluna]
    
    # Verificar se s√£o iguais
    if coluna_df1.equals(coluna_df2_base):
        print(f"   ‚úÖ Conte√∫do id√™ntico na coluna '{coluna}'")
    else:
        print(f"   ‚ö†Ô∏è Diferen√ßas encontradas na coluna '{coluna}'")
        conteudo_identico = False
        
        # Encontrar as diferen√ßas espec√≠ficas
        try:
            # Comparar linha por linha
            diferencas_linha = []
            for idx in range(min(len(coluna_df1), len(coluna_df2_base))):
                val1 = coluna_df1.iloc[idx]
                val2 = coluna_df2_base.iloc[idx]
                
                # Verificar se s√£o diferentes (considerando NaN)
                if pd.isna(val1) and pd.isna(val2):
                    continue  # Ambos s√£o NaN, consideramos iguais
                elif pd.isna(val1) or pd.isna(val2):
                    diferencas_linha.append(idx)
                elif val1 != val2:
                    diferencas_linha.append(idx)
            
            if diferencas_linha:
                diferencas_por_coluna[coluna] = diferencas_linha
                print(f"      üìä Total de linhas diferentes: {len(diferencas_linha)}")
                print(f"      üî∏ Primeiras 10 linhas diferentes: {diferencas_linha[:10]}")
                
                # Mostrar alguns exemplos das diferen√ßas
                print(f"      üìã Exemplos das diferen√ßas:")
                for i, linha_idx in enumerate(diferencas_linha[:5]):
                    val1 = coluna_df1.iloc[linha_idx]
                    val2 = coluna_df2_base.iloc[linha_idx]
                    print(f"         Linha {linha_idx}: '{val1}' vs '{val2}'")
                    
        except Exception as e:
            print(f"      ‚ö†Ô∏è Erro ao comparar coluna '{coluna}': {e}")
    
    # Estat√≠sticas b√°sicas da coluna
    print(f"      üìä Estat√≠sticas:")
    print(f"         - Valores √∫nicos DF1: {coluna_df1.nunique()}")
    print(f"         - Valores √∫nicos DF2: {coluna_df2_base.nunique()}")
    print(f"         - Valores nulos DF1: {coluna_df1.isnull().sum()}")
    print(f"         - Valores nulos DF2: {coluna_df2_base.isnull().sum()}")

# Resumo da verifica√ß√£o de conte√∫do
print(f"\nüéØ RESUMO DA VERIFICA√á√ÉO DE CONTE√öDO:")
print("="*50)
if conteudo_identico:
    print("‚úÖ CONTE√öDO ID√äNTICO: Todas as 12 colunas base t√™m conte√∫do id√™ntico!")
else:
    print("‚ö†Ô∏è CONTE√öDO DIFERENTE: Foram encontradas diferen√ßas nas seguintes colunas:")
    for coluna, linhas_diferentes in diferencas_por_coluna.items():
        print(f"   üî∏ {coluna}: {len(linhas_diferentes)} linhas diferentes")

# Verifica√ß√£o adicional: comparar DataFrames completos das colunas base
print(f"\nüîç VERIFICA√á√ÉO GLOBAL DAS COLUNAS BASE:")
try:
    # Resetar √≠ndices para garantir compara√ß√£o correta
    df1_reset = df1.reset_index(drop=True)
    df2_base_reset = df2_base.reset_index(drop=True)
    
    if df1_reset.equals(df2_base_reset):
        print("‚úÖ VERIFICA√á√ÉO GLOBAL: Os DataFrames das colunas base s√£o completamente id√™nticos!")
    else:
        print("‚ö†Ô∏è VERIFICA√á√ÉO GLOBAL: Os DataFrames das colunas base s√£o diferentes.")
        
        # Tentar usar compare() para mostrar diferen√ßas
        try:
            comparacao = df1_reset.compare(df2_base_reset)
            if not comparacao.empty:
                print(f"üìä Diferen√ßas encontradas em {len(comparacao)} posi√ß√µes:")
                print("üî∏ Primeiras 10 diferen√ßas:")
                print(comparacao.head(10))
            else:
                print("üîç Nenhuma diferen√ßa espec√≠fica encontrada pelo m√©todo compare()")
        except Exception as e:
            print(f"‚ö†Ô∏è Erro ao usar compare(): {e}")
            
except Exception as e:
    print(f"‚ö†Ô∏è Erro na verifica√ß√£o global: {e}")

print(f"\nüìã RESULTADO FINAL DA VERIFICA√á√ÉO DE CONTE√öDO:")
print(f"‚úÖ Conte√∫do das colunas base id√™ntico: {'Sim' if conteudo_identico else 'N√£o'}")

print(f"\nüîç COMPARA√á√ÉO DOS DADOS BASE:")
# Verificar se os dados das colunas base s√£o iguais
if df1.equals(df2_base):
    print("‚úÖ Os dados das colunas base s√£o id√™nticos.")
    dados_iguais = True
else:
    print("‚ö†Ô∏è Os dados das colunas base s√£o diferentes.")
    dados_iguais = False
    
    # Tentar encontrar as diferen√ßas espec√≠ficas
    try:
        diferencas = df1.compare(df2_base)
        if not diferencas.empty:
            print("üîç Diferen√ßas encontradas nas colunas base:")
            print(f"üìä Total de diferen√ßas: {len(diferencas)} linhas")
            print("üî∏ Primeiras 10 diferen√ßas:")
            print(diferencas.head(10))
        else:
            print("üîç Nenhuma diferen√ßa encontrada entre os dados base.")
    except Exception as e:
        print(f"‚ö†Ô∏è Erro ao comparar dados: {e}")
        
        # Compara√ß√£o alternativa linha por linha
        print("üîÑ Tentando compara√ß√£o alternativa...")
        linhas_diferentes = []
        for idx in range(min(len(df1), len(df2_base))):
            if not df1.iloc[idx].equals(df2_base.iloc[idx]):
                linhas_diferentes.append(idx)
                if len(linhas_diferentes) <= 10:  # Mostrar apenas as primeiras 10
                    print(f"   üî∏ Linha {idx}: Diferen√ßa encontrada")
        
        if linhas_diferentes:
            print(f"üìä Total de linhas diferentes: {len(linhas_diferentes)}")
        else:
            print("‚úÖ Nenhuma linha diferente encontrada")

# Verificar valores nulos nas colunas base
print(f"\nüîç AN√ÅLISE DE VALORES NULOS NAS COLUNAS BASE:")
nulos_df1 = df1.isnull().sum()
nulos_df2_base = df2_base.isnull().sum()

if nulos_df1.any() or nulos_df2_base.any():
    print("‚ö†Ô∏è Existem valores nulos nas colunas base:")
    if nulos_df1.any():
        print(f"üîπ Valores nulos no arquivo 1:\n{nulos_df1[nulos_df1 > 0]}")
    if nulos_df2_base.any():
        print(f"üîπ Valores nulos no arquivo 2 (colunas base):\n{nulos_df2_base[nulos_df2_base > 0]}")
else:
    print("‚úÖ Nenhum valor nulo encontrado nas colunas base.")

# Verificar duplicatas nas colunas base
print(f"\nüîç AN√ÅLISE DE DUPLICATAS NAS COLUNAS BASE:")
duplicatas_df1 = df1.duplicated().sum()
duplicatas_df2_base = df2_base.duplicated().sum()

if duplicatas_df1 > 0 or duplicatas_df2_base > 0:
    print("‚ö†Ô∏è Existem duplicatas nas colunas base:")
    print(f"üîπ Duplicatas no arquivo 1: {duplicatas_df1}")
    print(f"üîπ Duplicatas no arquivo 2 (colunas base): {duplicatas_df2_base}")
else:
    print("‚úÖ Nenhuma duplicata encontrada nas colunas base.")

# Verificar valores √∫nicos nas colunas base
print(f"\nüîç AN√ÅLISE DE VALORES √öNICOS NAS COLUNAS BASE:")
unicos_df1 = df1.nunique()
unicos_df2_base = df2_base.nunique()

print("üìä Compara√ß√£o de valores √∫nicos por coluna:")
for coluna in colunas_df1:
    if unicos_df1[coluna] != unicos_df2_base[coluna]:
        print(f"   üî∏ {coluna}: {unicos_df1[coluna]} vs {unicos_df2_base[coluna]}")

# Mostrar as colunas adicionais do df2 (colunas calculadas)
print(f"\nüîç COLUNAS ADICIONAIS NO ARQUIVO 2 (CALCULADAS):")
if len(df2.columns) > num_colunas_base:
    colunas_calculadas = list(df2.columns[num_colunas_base:])
    print(f"üìä Total de colunas calculadas: {len(colunas_calculadas)}")
    print(f"üîπ Colunas calculadas: {colunas_calculadas}")
    
    # Verificar se h√° valores nulos nas colunas calculadas
    df2_calculadas = df2.iloc[:, num_colunas_base:]
    nulos_calculadas = df2_calculadas.isnull().sum()
    if nulos_calculadas.any():
        print("‚ö†Ô∏è Valores nulos nas colunas calculadas:")
        print(f"{nulos_calculadas[nulos_calculadas > 0]}")
    else:
        print("‚úÖ Nenhum valor nulo nas colunas calculadas.")
else:
    print("‚ö†Ô∏è Nenhuma coluna adicional encontrada no arquivo 2.")

# Resumo final
print(f"\nüéØ RESUMO FINAL DA COMPARA√á√ÉO:")
print(f"‚úÖ Colunas base id√™nticas: {'Sim' if colunas_base_iguais else 'N√£o'}")
print(f"‚úÖ Tipos de dados id√™nticos: {'Sim' if tipos_iguais else 'N√£o'}")
print(f"‚úÖ Dados base id√™nticos: {'Sim' if dados_iguais else 'N√£o'}")
print(f"üìä Integridade dos dados base: {'‚úÖ APROVADA' if (colunas_base_iguais and tipos_iguais and dados_iguais) else '‚ö†Ô∏è REQUER ATEN√á√ÉO'}")
```

```python
import pandas as pd
from datetime import datetime
# Necess√°rio instalar chardet se n√£o estiver presente
%pip install chardet
import chardet

# Fun√ß√£o para detectar encoding
def detectar_encoding(arquivo):
    with open(arquivo, 'rb') as f:
        result = chardet.detect(f.read())
    return result['encoding']

# Carregar os arquivos com encoding correto
try:
    encoding_arq1 = detectar_encoding(arquivo1)
    encoding_arq2 = detectar_encoding(arquivo2)
    
    df1 = pd.read_csv(arquivo1, encoding=encoding_arq1)
    df2 = pd.read_csv(arquivo2, encoding=encoding_arq2)
    
    print("‚úÖ Arquivos carregados com sucesso!")
    print(f"Encoding detectado - Arquivo 1: {encoding_arq1}")
    print(f"Encoding detectado - Arquivo 2: {encoding_arq2}")
except Exception as e:
    print(f"‚ùå Erro ao carregar arquivos: {e}")
    raise

## 1. Padroniza√ß√£o de Datas ======================================

def converter_datas(df, colunas_data):
    for col in colunas_data:
        try:
            # Tentar converter do formato '01DEZ17 02:00'
            df[col] = pd.to_datetime(df[col], format='%d%b%y %H:%M', errors='coerce')
            
            # Se n√£o conseguir, tentar formato ISO
            if df[col].isnull().any():
                df[col] = pd.to_datetime(df[col], errors='coerce')
                
            print(f"‚úÖ Coluna {col} convertida para datetime")
        except Exception as e:
            print(f"‚ö†Ô∏è Erro ao converter coluna {col}: {e}")
    return df

# Colunas que precisam de convers√£o
colunas_datas = ['Checkin', 'Start', 'End', 'Checkout']

print("\nüîß Convertendo formatos de data no Arquivo 1...")
df1 = converter_datas(df1, colunas_datas)

print("\nüîß Convertendo formatos de data no Arquivo 2...")
df2 = converter_datas(df2, colunas_datas)

## 2. Corre√ß√£o de Encoding =======================================

def corrigir_encoding(texto):
    if isinstance(texto, str):
        try:
            # Corrigir caracteres mal codificados
            return texto.encode('latin1').decode('utf-8')
        except:
            return texto
    return texto

print("\nüîß Corrigindo encoding no Arquivo 2...")
df2['Crew'] = df2['Crew'].apply(corrigir_encoding)

# Verificar se ainda h√° diferen√ßas
dif_crew = df1['Crew'].compare(df2['Crew'])
print(f"\nüìä Diferen√ßas na coluna Crew ap√≥s corre√ß√£o: {len(dif_crew)}")

## 3. Verifica√ß√£o das Colunas Calculadas =========================

# Extrair colunas calculadas do df2
colunas_base = df1.columns.tolist()
colunas_calculadas = [col for col in df2.columns if col not in colunas_base]

print("\nüîç Analisando colunas calculadas:")
print(f"üìä Colunas calculadas encontradas: {colunas_calculadas}")

# Verificar valores nulos
print("\nüîç Valores nulos nas colunas calculadas:")
print(df2[colunas_calculadas].isnull().sum())

## 4. Compara√ß√£o Final ===========================================

print("\nüîé COMPARA√á√ÉO FINAL APOS PADRONIZA√á√ÉO")

# Fun√ß√£o para comparar colunas
def comparar_colunas(df1, df2, coluna):
    try:
        if df1[coluna].equals(df2[coluna]):
            return "‚úÖ Id√™ntico"
        else:
            dif = df1[coluna].compare(df2[coluna])
            return f"‚ö†Ô∏è Diferen√ßas: {len(dif)} linhas"
    except Exception as e:
        return f"‚ùå Erro: {str(e)}"

# Comparar cada coluna
print("\nüìã Compara√ß√£o coluna a coluna:")
for col in colunas_base:
    resultado = comparar_colunas(df1, df2, col)
    print(f"- {col}: {resultado}")

## 5. Salvando Resultados ========================================

# Criar dataframe com compara√ß√£o detalhada
resultados = []
for col in colunas_base:
    if df1[col].equals(df2[col]):
        resultados.append({'Coluna': col, 'Status': 'OK', 'Diferen√ßas': 0})
    else:
        dif = df1[col].compare(df2[col])
        resultados.append({'Coluna': col, 'Status': 'DIVERGENTE', 'Diferen√ßas': len(dif)})

df_resultados = pd.DataFrame(resultados)

# Adicionar estat√≠sticas
total_colunas = len(colunas_base)
colunas_ok = len(df_resultados[df_resultados['Status'] == 'OK'])
colunas_divergentes = total_colunas - colunas_ok

print("\nüéØ RESUMO FINAL:")
print(f"‚úÖ Colunas id√™nticas: {colunas_ok}/{total_colunas}")
print(f"‚ö†Ô∏è Colunas divergentes: {colunas_divergentes}/{total_colunas}")

# Mostrar colunas com problemas
if colunas_divergentes > 0:
    print("\nüîç Colunas com diverg√™ncias:")
    print(df_resultados[df_resultados['Status'] == 'DIVERGENTE'])

# Salvar relat√≥rio
df_resultados.to_csv('relatorio_comparacao.csv', index=False)
print("\nüìÑ Relat√≥rio salvo como 'relatorio_comparacao.csv'")

# Salvar vers√µes padronizadas
df1.to_csv('arquivo1_padronizado.csv', index=False)
df2.to_csv('arquivo2_padronizado.csv', index=False)
print("üíæ Arquivos padronizados salvos:")
print("- arquivo1_padronizado.csv")
print("- arquivo2_padronizado.csv")
```
